<h2><strong>Marco teórico</strong></h2>
<hr style="height:2px;border-width:0;color:gray;background-color:violet">
<p id='marco_t' style="text-align:justify; font-size:13px"> 
$\quad$ Si bien, hay variables que intuitivamente se espera que estén correlacionadas, hay evidencia empírica de que algunas de nuestras variables de interés están correlacionadas como es el caso del <em>PIB</em> y la expectativa de vida <sup><a href="#1">[1]</a></sup> <sup><a href="#2">[2]</a></sup> <sup><a href="#3">[3]</a></sup> cuya relación analítica está dada por la siguiente ecuación
</p>

$$
Expectativa \,\, de \,\, vida = \beta_0 + \beta_1 \log\left( PIB \right) + u_t,
$$

<p style="text-align:justify; font-size:13px"> 
donde los errores $u_t$ están identicamente distribuidos con una distribución normal estándar. A pesar de que esta relación se aproxime mejor para países con un <em>PIB</em> bajo, por lo general se toma para todos los países puesto que la correlación tiende a ser bastante significativa. Por otro lado, a pesar de que el <em>PIB</em> sea utilizado como una medida del bienestar, no necesariamente implica que un país con un <em>PIB</em> grande tenga un alto índice de calidad de vida. Sin embargo, cabe preutnarse si estas variables están correlacionadas significativamente para la mayoría de los países, en particular, veremos si existe alguna relación significativa entre esta y otras variables.
Análogamente, las variables que estén suficientemente correlacionadas serán regresadas entre sí mientras no haya multicolinealidad, aunque hay variables que se esperan estar correlacioandas como el <em>PIB</em> y la <em>tasa de inflación</em><sup><a href="#11">[2]</a></sup>. Sin embargo, también hay posibles relaciones intuitivas tales como una relación a nivel general entre la <em>expectativa de vida</em> y la <em>calidad de vida</em> a pesar de que haya países que sirvan de contraejemplo tales como China. Cabe mencionar que no se descarta que hayan variables correlacionadas que no tengan una relación causal o explícita más allá de los datos, pero que sirva para futuras comparativas en otros trabajos.
<br>
<br>
$\quad$ Por otra parte, se aplicarán técnicas de <em>clustering</em> <sup><a href="#7">[4]</a></sup> para detectar la cantidad de grupos representativos óptimos por medio del algoritmo llamado <em>K-means++</em> <sup><a href="#4">[5]</a></sup>. Este algoritmo está basado en el agoritmo usual del <em>K-means</em> con la diferencia de que este algoritmo resuelve la problemática <em>NP-duro k-means</em> donde el agoritmo usual arrojaba centros de clusters que satisfacían la condición de mínima distancia del algoritmo, pero que visualmente se apreciaba que los culsters están incorretamente definidos. Por otro lado, en este trabajo no se usará un conjunto de etiquetas discretas, sino un conjunto contínuo el cual será mapeado a un gradiente de color que nos permita visualizar tendencias o grupos representativos a partir de aplicar un umbral para determinar el número de cluster necesarios para cada caso.
<br>
<br>
    En este trabajo se omitirán los <em>outliers</em>, a excepción de la <em>Población</em>, para el correcto funionamiento de los algorítmos de clasificación  regresión. El método que se utilizará será mediante el uso del cálculo de los cuartiles para filtrar fuera los datos que no estén dentro del segundo y tercer cuartil más 1.5 veces el rango intercuartil de los datos correspondientes <sup><a href="#12">[6]</a></sup>.
<br>
<br>
Debido al mes en el que se está haciendo este proyecto, las bases de datos del año en curso, 2022, no han sido propiamente actualizadas, no están completas o las fuentes no son confiables; es por este motivo que para este análisis nos restringiremos a los datos del año 2021.

Añadamos el intercepto para la regresión

```python
dataframe['const'] = 1
```

```python
reg1 = sm.OLS(endog = dataframe['Expectativa_vida'], 
              exog = dataframe[['const', 'log_PIB_pc']],
              missing='drop')
type(reg1)
```

```python
results = reg1.fit()
type(results)
```

```python
print(results.summary())
```

```python
dataframe_subset = dataframe.dropna(subset=['const', 'log_PIB_pc'])
```

```python
mean_expr = np.mean(dataframe_subset['log_PIB_pc'])
mean_expr
```

```python
results.predict(exog=[1, mean_expr])
```

```python
# Plot predicted values

fix, ax = plt.subplots()
ax.scatter(dataframe_subset['log_PIB_pc'], results.predict(), alpha=0.5,
        label='predicted')

# Plot observed values

ax.scatter(dataframe_subset['log_PIB_pc'], dataframe_subset['Expectativa_vida'], 
           alpha=0.5, label='observed')

ax.legend()
ax.set_title('OLS predicted values')
ax.set_xlabel('log_PIB_pc')
ax.set_ylabel('Expectativa_vida')
plt.show()
```

```python
# Create lists of variables to be used in each regression
X1 = ['const', 'log_PIB_pc']
X2 = ['const', 'Inflacion']
X3 = ['const', 'Calidad_de_vida']

# Estimate an OLS regression for each set of variables
reg1 = sm.OLS(dataframe['Expectativa_vida'], dataframe[X1], missing='drop').fit()
reg2 = sm.OLS(dataframe['Expectativa_vida'], dataframe[X2], missing='drop').fit()
reg3 = sm.OLS(dataframe['Expectativa_vida'], dataframe[X3], missing='drop').fit()
```

```python
info_dict={'R-squared' : lambda x: f"{x.rsquared:.2f}",
           'No. observations' : lambda x: f"{int(x.nobs):d}"}

results_table = summary_col(results=[reg1,reg2,reg3],
                            float_format='%0.2f',
                            stars = True,
                            model_names=['Model 1',
                                         'Model 3',
                                         'Model 4'],
                            info_dict=info_dict,
                            regressor_order=['const',
                                             'log_PIB_pc',
                                             'Inflacion',
                                             'Calidad_de_vida'])

results_table.add_title('Table 2 - OLS Regressions')

print(results_table)
```

```python
# Para rellenar los datos faltantes con el promedio, se sacará primero el promedio por columna
# luego, se mete este valor para usarse en el método de *fillna*
promedio_pib = df_regresion['log_PIB_pc'].mean(skipna = True)
promedio_expectativa = df_regresion['Expectativa_vida'].mean(skipna = True)
promedio_inflacion = df_regresion['Inflacion'].mean(skipna = True)
promedio_calidad = df_regresion['Calidad_de_vida'].mean(skipna = True)

df_regresion['log_PIB_pc'].fillna(value = promedio_pib, inplace = True)
df_regresion['Expectativa_vida'].fillna(value = promedio_expectativa, inplace = True)
df_regresion['Inflacion'].fillna(value = promedio_inflacion, inplace = True)
df_regresion['Calidad_de_vida'].fillna(value = promedio_calidad, inplace = True)
print("Valores faltantes del dataframe: " + str(df_regresion.isna().sum().sum()))
```

Adicionalmente a las gráficas en el conjunto de prueba, se evaluará el desempeño de ambos modelos por medio de validación cruzada y matrices de confusión, respectivamente. Por otro lado, para el caso de segmentación de intereés por el nivel del PIB se aplicará el meodo del codo para corroborar que el número de clústeres propuestos es correcto, este meodo usará KMeans++ para este método ya que se especializada en clusterización de manera precisa y eficaz-

```python
# Conjuntos de datos de Expectativa de vida vs Log(PIB pc)
X_lr_1 = df_regresion.iloc[:, 1:2].values
y_lr_1 = df_regresion.iloc[:, 3:4].values
# Conjuntos de datos de Log(PIB pc) vs Índice de caldiad de vida
X_lr_2 = df_regresion.iloc[:, 4:5].values
y_lr_2 = df_regresion.iloc[:, 1:2].values
# Conjuntos de datos de Inflación vs log(PIB pc)
X_lr_3 = df_regresion.iloc[:, 1:2].values
y_lr_3 = df_regresion.iloc[:, 2:3].values
```

```python
df_regresion.head()
```

```python
# CRearemos listas que guarden los modelos y arreglos generados
Num_reg = 3
X_lr = [X_lr_1, X_lr_2, X_lr_3]
y_lr = [y_lr_1, y_lr_2, y_lr_3]
# Conjuntos del split
X_lr_train_list = [None]*Num_reg
y_lr_train_list = [None]*Num_reg
X_lr_test_list = [None]*Num_reg
y_lr_test_list = [None]*Num_reg
y_lr_pred_list = [None]*Num_reg
# Precisiones
accuracy_scores_train_lr = np.ones((Num_reg))
accuracy_scores_test_lr = np.ones((Num_reg))
# Coeficientes de la regresión
interceptos = [None]*Num_reg
ordenada = [None]*Num_reg
# Modelos
regresion_list_lr = [None]*Num_reg
```

```python
# Aplicaremos los algoritmos vistos en clase, pero en buvle sobre las listas que creamos
for i in range(Num_reg):
    X_lr_train_list[i], X_lr_test_list[i], y_lr_train_list[i], y_lr_test_list[i] = train_test_split(X_lr[i], y_lr[i], 
                                                                                                    test_size = 0.2, 
                                                                                                    random_state = 0)
    
    regresion_list_lr[i] = LinearRegression()
    regresion_list_lr[i].fit(X_lr_train_list[i], y_lr_train_list[i])
    
    accuracy_scores_train_lr[i] = regresion_list_lr[i].score(X_lr_train_list[i], y_lr_train_list[i])
    accuracy_scores_test_lr[i] = regresion_list_lr[i].score(X_lr_test_list[i], y_lr_test_list[i])
    
    y_lr_pred_list[i] = regresion_list_lr[i].predict(X_lr_test_list[i])
```

```python
tittles_train = ['Conjunto de entrenamiento']*Num_reg
tittles_test = ['Conjunto de prueba']*Num_reg
labels_y = ['Expectativa de vida [años]', 'Índice de calidad de vida', 'Inflación [$\$$]']
```

```python
# La lógica de esta gráfica es análoga a anteriores con grid o subplots
fig, axes = plt.subplots(3, 2, figsize=(20,20), 
                         sharey = False, sharex = False)
fig.suptitle("Regresiones lineales obtenidas", 
             fontsize = 45, color = 'darkviolet', 
             fontweight = 'bold', y = 0.94)
fig.patch.set_facecolor('mintcream')
fig.patch.set_alpha(0.99)

for i in range(Num_reg):
    # Conjunto de entrenamiento
    axes[i,0].scatter(X_lr_train_list[i], y_lr_train_list[i], 
                      color = "magenta", s = 50)
    axes[i,0].plot(X_lr_train_list[i], regresion_list_lr[i].predict(X_lr_train_list[i]), 
                   color = "black", linewidth = 4)
    axes[i,0].set_title(tittles_train[i], 
                        fontsize = 19, fontweight = 'bold', 
                        color = 'purple')
    axes[i,0].set_xlabel('Log(PIB per cápita)',
                         fontsize = 16, color = 'navy',
                         fontweight = 'bold')
    axes[i,0].set_ylabel(labels_y[i], 
                         fontsize = 16, color = 'navy',
                         fontweight = 'bold')
    axes[i,0].tick_params(labelsize = 12,
                          labelcolor = 'maroon')
    axes[i,0].grid(color = 'royalblue', 
                   linestyle = '-', alpha = 1)
    axes[i,0].patch.set_facecolor('cyan')
    axes[i,0].patch.set_alpha(0.8)
    
    # Conjunto de prueba
    axes[i,1].scatter(X_lr_test_list[i], y_lr_test_list[i], 
                      color = "cyan", s = 50)
    axes[i,1].plot(X_lr_train_list[i], regresion_list_lr[i].predict(X_lr_train_list[i]), 
                   color = "black", linewidth = 4,
                   linestyle = ':')
    axes[i,1].set_title(tittles_test[i], 
                        fontsize = 19, fontweight = 'bold', 
                        color = 'purple')
    axes[i,1].set_xlabel('Log(PIB per cápita)',
                         fontsize = 16, color = 'navy',
                         fontweight = 'bold')
    axes[i,1].set_ylabel(labels_y[i], 
                         fontsize = 16, color = 'navy',
                         fontweight = 'bold')
    axes[i,1].tick_params(labelsize = 12,
                          labelcolor = 'maroon')
    axes[i,1].grid(color = 'yellow', 
                   linestyle = '-', alpha = 1)
    axes[i,1].patch.set_facecolor('magenta')
    axes[i,1].patch.set_alpha(0.8)

plt.subplots_adjust(wspace = 0.18, hspace = 0.26)
plt.savefig(path + "Proyecto_E_II/linear_models.png")    
plt.show()
```
