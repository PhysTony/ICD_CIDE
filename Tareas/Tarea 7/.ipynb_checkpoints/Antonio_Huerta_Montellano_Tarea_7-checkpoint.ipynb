**(i)** Cargar el archivo tips y hacer un histograma con la variable 'total_bill' traslapado con una función de densidad de los datos, usar seaborn.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
print("Modulos importados")
```

```python
print("Para modificar el tamaño de todos los gráficos")
from matplotlib import rcParams
rcParams['figure.figsize'] = 15,9
```

```python
print("Para que no nos molesten los mensajes de peligro.")
import warnings
warnings.filterwarnings('ignore')
```

```python
tips = pd.read_csv("tips.csv")
tips.head()
```

Veamos el mínimo y máximo de la columna *total_bill* para saber cuántos bins tomar.

```python
print(tips['total_bill'].min())
print(tips['total_bill'].max())
```

Podemos tomar 9 bins como aproximación.

```python
sns.distplot(tips['total_bill'], bins=9,
             hist_kws={'color':'#DC143C', 'edgecolor':'#aaff00',
                       'linewidth':5, 'linestyle':'--',
                       'alpha':0.85},
             kde_kws={"color": "#8e00ce", "linewidth": 3})
plt.title("Total bill histogram and density function", fontsize = 20, fontweight = 'bold')
plt.xlabel("Total bill[$\$$]", fontsize = 16, fontweight = 'bold')
plt.ylabel("Density", fontsize = 16, fontweight = 'bold')
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.grid()
plt.show()
```

**(ii**) Con el comando countplot de seaborn haz un diagrama de barras indicando el número de observaciones por día de la semana, poner título y leyendas en los ejes.

```python
sns.countplot(data = tips, x = 'day', linewidth = 3.5,
              palette = sns.color_palette("colorblind"),
              edgecolor = '#aaff00')
plt.title("Day histogram", fontsize = 20, fontweight = 'bold')
plt.xlabel("Day", fontsize = 16, fontweight = 'bold')
plt.ylabel("Frequanty", fontsize = 16, fontweight = 'bold')
plt.xticks(fontsize=16)
plt.yticks(fontsize=14)
plt.grid(color='gray', linestyle='--')
plt.show()
```

**(iii)** Con el comando regplot de seaborn haz un diagrama de dispersión de las variables 'total_bill' y 'tip', presenta la línea de regresión ajustada, los titulos y leyendas en los ejes.

```python
sns.regplot(x = 'tip', y = 'total_bill', data = tips,
            scatter_kws={"color": '#DC143C', 'edgecolor':'#CCCC00', "s":75, "alpha":0.75},
            line_kws={"color": "#8e00ce"})
plt.title("Tip vs Total bill", fontsize = 20, fontweight = 'bold')
plt.xlabel("Total bill[$\$$]", fontsize = 16, fontweight = 'bold')
plt.ylabel("Tip [$\$$]", fontsize = 16, fontweight = 'bold')
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.xlim(0.9,10.1)
plt.grid(color='purple', linestyle='-')
plt.show()
```

**(iv)** Con el comando kdeplot de seaborn haz una gráfica de densidad en 2D para las variables  'total_bill' y 'tip', presenta el titulo y leyenda en los ejes.

```python
f , axes = plt.subplots(1,1)
axes.patch.set_facecolor("#030210")
axes.patch.set_alpha(0.9)
relleno = sns.kdeplot(x = 'total_bill', y = 'tip', data = tips,
                      shade = True , shadelowest = True,
                      cmap = 'inferno')
contorno = sns.kdeplot(x = 'total_bill', y = 'tip', data = tips,
                       cmap = 'cool')
plt.title("Tip vs Total bill KDE - Forma 1", fontsize = 20, fontweight = 'bold')
plt.xlabel("Total bill[$\$$]", fontsize = 16, fontweight = 'bold')
plt.ylabel("Tip [$\$$]", fontsize = 16, fontweight = 'bold')
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.grid(color = 'purple', linestyle = '-')
plt.show()
```

```python
sns.kdeplot(x = 'total_bill', y = 'tip', data = tips,
            fill=True, thresh=0, levels=100, cmap="mako")
plt.title("Tip vs Total bill KDE - Forma 2", fontsize = 20, fontweight = 'bold')
plt.xlabel("Total bill[$\$$]", fontsize = 16, fontweight = 'bold')
plt.ylabel("Tip [$\$$]", fontsize = 16, fontweight = 'bold')
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.grid(visible = None)
plt.show()
```

**(v)** Con el comando boxplot de seaborn presenta la mediana, cuartiles y outliers para la  variable 'total_bill' de acuerdo con el 'time', presenta el titulo y leyenda en los ejes.

```python
PROPS = {
    'boxprops':{'facecolor':'#DC143C','edgecolor':'#aaff00', 'linewidth':5, 'alpha':0.7},
    'medianprops':{'color':'#8e00ce', 'linewidth':3},
    'whiskerprops':{'color':'#aaff00'},
    'flierprops' :{'marker':'o', 'markerfacecolor':'magenta', 'markersize':12, 'markeredgecolor':'cyan'},
    'capprops':{'color':'#8e00ce'}
}
sns.boxplot(x = 'time', y = 'total_bill', data = tips,
            order = ['Lunch', 'Dinner'],
            palette = 'Set2', **PROPS)
plt.title("Total bill boxplot", fontsize = 20, fontweight = 'bold')
plt.xlabel("Time", fontsize = 16, fontweight = 'bold')
plt.ylabel("Total bill frequenty", fontsize = 16, fontweight = 'bold')
plt.xticks(fontsize=14)
plt.yticks(fontsize=16)
plt.grid(color='gray', linestyle='--')
plt.show()
```

**(vi)** Con el comando violinplot de seaborn presenta la distribución que tienen los datos de la variable 'total_bill' de acuerdo con el 'time', presenta el titulo y leyenda en los ejes.

```python
sns.violinplot(x = 'time', y = 'total_bill', data = tips,
              linewidth = 3, inner = None,
              palette = 'pastel', edgecolor = '#aaff00')
sns.boxenplot(x = 'time', y = 'total_bill', data = tips,
              color="red", width=0.025)
plt.title("Total bill violinplot", fontsize = 20, fontweight = 'bold')
plt.xlabel("Time", fontsize = 16, fontweight = 'bold')
plt.ylabel("Total bill frequenty", fontsize = 16, fontweight = 'bold')
plt.xticks(fontsize=14)
plt.yticks(fontsize=16)
plt.grid(color='gray', linestyle='--')
plt.show()
```

**(vii)** Con el comando pairplot de seaborn, presenta histogramas para las tres variables 'tip', 'total_bill' y 'size', asi como todas las combinaciones de diagramas de dispersión.

```python
colors = iter(['xkcd:red purple', 'xkcd:pale teal', 'xkcd:warm purple',
       'xkcd:light forest green', 'xkcd:blue with a hint of purple',
       'xkcd:light peach', 'xkcd:dusky purple', 'xkcd:pale mauve',
       'xkcd:bright sky blue', 'xkcd:baby poop green', 'xkcd:brownish',
       'xkcd:moss green', 'xkcd:deep blue', 'xkcd:melon',
       'xkcd:faded green', 'xkcd:cyan', 'xkcd:brown green',
       'xkcd:purple blue', 'xkcd:baby shit green', 'xkcd:greyish blue'])
# Fuera de la diagonal:
def my_scatter(x,y, **kwargs):
    kwargs['color'] = next(colors)
    kwargs['alpha'] = 0.2
    plt.scatter(x,y, **kwargs)
# En la diagonal:
def my_hist(x, **kwargs):
    kwargs['color'] = next(colors)
    kwargs['alpha'] = 0.2
    kwargs['linewidth'] = 0.2
    plt.hist(x, **kwargs)
```

```python
pp = sns.pairplot(tips)
pp.map_diag(my_hist)
pp.map_offdiag(my_scatter)
pp.fig.suptitle("Pairplot", fontsize = 24, color = 'magenta', fontweight = 'bold', y=1.08)
pp.fig.set_size_inches(15,15)
plt.xticks(fontsize=14)
plt.yticks(fontsize=16)
plt.show()
```
