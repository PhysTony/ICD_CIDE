## 7. Programacion orientada a arreglos

### Por lo general las operaciones con arreglos vectorizados son mucho más rápidas que las que se pueden realizar con otros objetos de Python. 
### Ejemplo:  se quiere realizar la operación sqrt(x^2  +   y^2) en los escalares de un arreglo

```python
points = np.arange(-5, 5, 0.01)               # Se crean puntos igualmente espaciados de -5 a 5 con diferencias de .01
points
```

```python
xs, ys= np.meshgrid(points, points)           # Creamos dos matrices a partir del arreglo unidimensional
xs
```

```python
ys
```

```python
z = np.sqrt(xs ** 2 + ys ** 2)                # Escribimos la función matemática a aplicar como si se tratara de dos escalares
z
```

### Como un adelanto del proceso de visualizaciones creamos con matplotlib unas gráficas

```python
import matplotlib.pyplot as plt
plt.imshow(z, cmap=plt.cm.gray)   # con imshow creamos un mapa de calor en el que la intensidad de gris refleja la función 
plt.colorbar()                    # establecemos una leyenda con tonos de gris y valores correspondientes
plt.title("Image plot of $\sqrt{x^2 + y^2}$ for a grid of values")       # Titulo de la gráfica
```

### Las operaciones lógicas se facilitan con arreglos vectorizados

```python
xarr = np.array([1.1, 1.2, 1.3, 1.4, 1.5])
yarr = np.array([2.1, 2.2, 2.3, 2.4, 2.5])
cond = np.array([True, False, True, True, False])
```

```python
result = [(x if c else y) for x, y, c in zip(xarr, yarr, cond)]      # Aplicamos a los arreglos una instrucción comprimida
result                             # Vamos a elegir el valor de x si c es Verdadero, y el de y si c es Falso
```

```python
result = np.where(cond, xarr, yarr)               # con el comando where de Numpy la instrucción es más sencilla
result                                            # El resultado se obtiene mucho más rápido cuando se trata de arreglos grandes
```

## 11. Ejemplo: Caminata aleatoria

### En una caminata aleatoria sencilla se inicia con el valor 0 y se procede con valores de 1 y -1 que ocurren con la misma probabilidad

```python
import random                               # importamos la librería
position = 0
walk = [position]                           # en la lista walk almacenamos las posiciones
steps = 1000
for i in range(steps):
   if random.randint(0, 1):                  # Genera un 0 o un 1 al azar, un 0 corresponde a un valor falso 
       step = 1 
   else: 
       step = -1
   position += step
   walk.append(position)
```

```python
plt.plot(walk[:100])                         # al poner los índices de la lista los grafica en el eje horizontal
```

### Como la caminata es simplemente la suma acumulada de los pasos aleatorios puedo utilizar una expresión de un arreglo

```python
nsteps = 1000
draws = np.random.randint(0, 2, size=nsteps) # se crea un arreglo con 0, 1 al azar de tamaño 1,0000            
steps = np.where(draws > 0, 1, -1)     # Se crea un arreglo; a los valores 1 se les pone 1 y a los de 0 se les pone -1 
walk = steps.cumsum()                             # se van sumando los valores de forma acumulada
```

```python
walk.min()
```

```python
walk.max()
```

### Si quremos simular varias caminatas al mismo tiempo

```python
nwalks = 5000
nsteps = 1000
draws = np.random.randint(0, 2, size=(nwalks, nsteps)) # Se crea una matriz de 0 o 1, las distintas caminatas son los renglones
steps = np.where(draws > 0, 1, -1)
walks = steps.cumsum(1)
walks
```

### Queremos calcular cuantas veces se rebasaron los valores -30 y  30

```python
hits30 = (np.abs(walks) >= 30).any(1)           # el metodo any checa si al menos uno de los elementos es Verdadero = 1
```

```python
hits30
```

```python
hits30.sum()
```
