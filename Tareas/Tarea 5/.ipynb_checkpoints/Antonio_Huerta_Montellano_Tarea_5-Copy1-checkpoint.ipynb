1. Explicar cómo en una red real se puede saber si se presentan las siguientes propiedades: trayectorias cortas, aglutinamiento, y jerarquía en los grados.
**Primero, veamos otra vez las definiciones de los algunos de los términos involucrado:**
- **_Longitud promedio de la trayectoria_ = indicador se obtiene con el número mínimo de vínculos indirectos entre dos nodos cualesquiera de la red.**
- **_Aglutinamiento_ = probabilidad de que dos nodos vinculados a un tercero estén conectados entre si**
**Tomemos en cuenta que una red real será finita, por lo que el aglutinamiento se podrá calcular de manera frecuentista ya que identificando las conexiones, se podrá determinar cuáles de los que están vinculados a un tercero también lo están. Por otro lado, como es una red finita, se podrá medir la longitud promedio de la trayectoria entre los elementos de la red y dependiendo del umbral para determinarse mínimo, se determinarán las trayectorias cortas. Por último, con respecto a la jerarquía, se puede definir como aquellos nodos que, a partir de cierto umbral, tienen más vínculos directos de una longitud promedio de la trayectoria igual a uno que el promedio.**

```python
import pandas as pd
import numpy as np
```

```python
data = {'year': [2010, 2011, 2012, 2010, 2011, 2012, 2010, 2011, 2012],
        'team': ['FCBarcelona', 'FCBarcelona', 'FCBarcelona', 'RMadrid', 'RMadrid', 'RMadrid', 'ValenciaCF', 'ValenciaCF', 'ValenciaCF'], 
        'wins': [30, 28, 32, 29, 32, 26, 21, 17, 19], 
        'draws': [6, 7, 4, 5, 4, 7, 8, 10, 8],
        'losses': [2, 3, 2, 4, 2, 5, 9, 11, 11]}
```

```python
# 1
dataframe = pd.DataFrame(data)
```

```python
# 9: Notemos que sacará el máximo de todos los valores independientemente de si son números o no.
def f_max(x):
    return pd.Series([x.max()], index=['max'])
dataframe.apply(f_max)
```

```python
# 9
f_max = lambda x: x.max()
dataframe.apply(f_max)
```

```python
# 10
dataframe[['wins']].min()
```

```python
# 11:
f_cuadrado = lambda x: x**2 
dataframe[['draws']].apply(f_cuadrado)
```

```python
# 11 Otra forma es creando una columna aparte que contenga dicho valor
dataframe[['draws$^{2}$']] = dataframe[['draws']].apply(f_cuadrado)
dataframe.head()
```

```python
# Eliminemos dicha columna
del dataframe['draws$^{2}$']
dataframe.head()
```

```python
# 12
dataframe['penalties'] = np.nan
dataframe.head()
```
