### Figuras y páneles (subplots) 

```python
fig = plt.figure()                # Creamos una nueva figura (visualización) 
```

```python
ax1 = fig.add_subplot(2, 2, 1)   # Establecemos la posiblidad de que tenga hasta 4 páneles 2 x 2 
ax2 = fig.add_subplot(2, 2, 2)   # El tercer argumento hace referencia al número de la figura
ax3 = fig.add_subplot(2, 2, 3)   #  Aquí creamos solo tres figuras
```

```python
fig = plt.figure()                # Escribimos todas las instrucciones de corrido
ax1 = fig.add_subplot(2, 2, 1)   
ax2 = fig.add_subplot(2, 2, 2)   
ax3 = fig.add_subplot(2, 2, 3)
plt.plot(np.random.randn(50).cumsum(), 'k--')    # Comando de graficación para la última figura
                                                 # cuando el panel no se espespecifica: ax1, ax2 o ax3
```

```python
fig = plt.figure()                # Le agregamos contenido a las tres graficas
ax1 = fig.add_subplot(2, 2, 1)   
ax2 = fig.add_subplot(2, 2, 2)   
ax3 = fig.add_subplot(2, 2, 3)    # Ahora vamos a especificar los páneles correspondientes
ax3.plot(np.random.randn(50).cumsum(), 'k--')    # Una serie de tiempo  (k es el color negro)
ax1.hist(np.random.randn(100), bins=20, color='k', alpha=0.3)     # Un histograma, alpha es el tono
ax2.scatter(np.arange(30), np.arange(30) + 3 * np.random.randn(30))   # Un diagrama de dispersión
```

```python
fig, axes = plt.subplots(2, 3)      # Otra forma de crear una visualización: 2 x 3 = 6 páneles
```

```python
fig, axes = plt.subplots(2, 3)      # Otra forma de crear una visualización: 2 x 3 = 6 páneles
plt.subplots_adjust(wspace=0.6, hspace=0.6) 
```

```python
# Espacios entre figuras
```

```python
fig, axes = plt.subplots(2, 2, sharex=True, sharey=True)
for i in range(2):                #sharex  = todos los paneles comparten mismos tics en eje x
    for j in range(2):                 # Creamos los cuatro histogramas
       axes[i, j].hist(np.random.randn(500), bins=50, color='k', alpha=0.5)
    plt.subplots_adjust(wspace=0, hspace=0)   # No dejamos espacio entre los distintos páneles
```

### Colores, marcadores y estilos de líneas

```python
from numpy.random import randn        # Importamos generador de numeros aleatorios
```

```python
plt.plot(randn(30).cumsum(), 'ko--') # Definimos que los puntos sean de color negro y línea segmentada
```

```python
plt.plot(randn(30).cumsum(), color='k',linestyle='dashed',marker='o') # Misma instrucción, más detalle
```

```python
data = np.random.randn(30).cumsum()        # Podemos cambiar el estilo, poniendo pasos
plt.plot(data, 'k-', drawstyle='steps-post', label='steps-post')
plt.legend(loc='best')                     # Agregamos además una leyenda 
```

```python
plt.plot(data, 'g--', label='Default')       # Podemos encimar dos tipos de estilos
plt.plot(data, 'k-', drawstyle='steps-post', label='steps-post')
plt.legend(loc='best') 
```

### Tics, etiquetas y leyendas

```python
fig = plt.figure()                         # Empecemos con una gráfica sencilla
ax = fig.add_subplot(1, 1, 1)
ax.plot(np.random.randn(1000).cumsum())
```

```python
fig = plt.figure() 
ax = fig.add_subplot(1, 1, 1)
ticks = ax.set_xticks([0, 250, 500, 750, 1000])   # Especificamos los tics que deseamos marcar
labels = ax.set_xticklabels(['one', 'two', 'three', 'four', 'five'],
rotation=30, fontsize='small')              # Si queremos les damos un nombre a los tics   
ax.set_title('My first matplotlib plot')    # Indicamos un título
ax.set_xlabel('Stages')                     # También una leyenda en el eje de las x
ax.plot(np.random.randn(1000).cumsum())
```

```python
# Agregar varias leyendas
```

```python
fig = plt.figure(); ax = fig.add_subplot(1, 1, 1)        # Un solo panel          
ax.plot(randn(1000).cumsum(), 'g', label='one')  # Creamos una primera serie de aleatorios acumulados
ax.plot(randn(1000).cumsum(), 'k--', label='two')  # Una segunda serie
ax.plot(randn(1000).cumsum(), 'b.', label='three') # Una tercera serie
ax.legend(loc='right')                              # Queremos que se distingan con una leyenda
plt.savefig('figura.png', dpi=400, bbox_inches='tight') 
# El archivo se puede guardar en el dasboard de Jupyter o en el disco si se le direcciona
# Puede usarse otro formato como PDF y con distintas resoluciones
```

### Integración de dibujos y comentarios en los paneles

```python
from datetime import datetime                # Veamos un ejemplo
import pandas as pd
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
data = pd.read_csv('spx.csv', index_col=0, parse_dates=True) # Importamos los datos de un archivo CSV
spx = data['SPX']                  # A partir de la estructura de datos 'data' asignamos una variable 
spx.plot(ax=ax, style='k-')
crisis_data = [                                  # En un iterador specificamos los tics (periodos) 
(datetime(2007, 10, 11), 'Peak of bull market'), # en los que se pone un letrero
(datetime(2008, 3, 12), 'Bear Stearns Fails'),
(datetime(2008, 9, 15), 'Lehman Bankruptcy')
]
for date, label in crisis_data:                           # Tomamos valores del iterador
   ax.annotate(label, xy=(date, spx.asof(date) + 75),     # posición y tamaño de letras
   xytext=(date, spx.asof(date) + 225),                   # posición y tamaño de la flechas
   arrowprops=dict(facecolor='red', headwidth=4, width=2, # características de la flecha
   headlength=4),
   horizontalalignment='left', verticalalignment='top')
# Zoom in on 2007-2010                    # Solo queremo graficar un periodo de la serie
ax.set_xlim(['1/1/2007', '1/1/2011'])     # Establecemos limites en eje de las x
ax.set_ylim([600, 1800])                  # Establecemos límites en eje de las y
ax.set_title('Important dates in the 2008-2009 financial crisis')
```

```python
# Incorporacion de algunas formas geométricas
```

```python
fig = plt.figure()                    # Otras formas disponibles en matplotlib.patches
ax = fig.add_subplot(1, 1, 1)
rect = plt.Rectangle((0.2, 0.75), 0.4, 0.15, color='k', alpha=0.3)
circ = plt.Circle((0.7, 0.2), 0.15, color='b', alpha=0.3)
pgon = plt.Polygon([[0.15, 0.15], [0.35, 0.4], [0.2, 0.6]],
color='g', alpha=0.5)
ax.add_patch(rect)
ax.add_patch(circ)
ax.add_patch(pgon)
```

```python
# Creación de un diagrama de frecuencia desde datos originales
```

```python
tips = pd.read_csv('tips.csv')       # Cargamos los datos desde el archivo tips
```

```python
# Archivo tiene número de comensales en cada grupo en cada día de la semana, entre otras variables 
```

```python
tips.head()
```

```python
party_counts = pd.crosstab(tips['day'], tips['size'])    # Hacemos tabulación cruzada para frecuencias
party_counts
```

```python
# Normalizamos a lo largo de rengló, para que la suma de la unidad
```

```python
party_pcts = party_counts.div(party_counts.sum(1), axis=0)   # la suma se calcula para cada renglón
party_pcts.plot.bar()               # Hacemos diagramas de barras con estos datos
```

```python
# Con la paquetería seaborn el asunto puede ser más simple
```

```python
import seaborn as sns   #Calculamos porcentaje del tip con respecto a cuenta neta (sin tip)
tips['tip_pct'] = tips['tip'] / (tips['total_bill'] - tips['tip'])  
tips.head()
```

```python
sns.barplot(x='tip_pct', y='day', data=tips, orient='h')  #Obtenemos promedio en dia e intervalo 95%
```

```python
sns.set(style="whitegrid")                  # Si queremos poner líneas reticulares
sns.barplot(x='tip_pct', y='day', hue='time', data=tips, orient='h')   # Si queremos acotarlos a hora
```

### Histogramas y diagramas de densidad

```python
tips['tip_pct'].plot.hist(bins=50)
```

```python
tips['tip_pct'].plot.density()  #Se ajusta una distribución continua (kernel density plots)
```

```python
comp1 = np.random.normal(0, 1, size=200)     # Podemos combinar histograma y densidad
comp2 = np.random.normal(10, 2, size=200)
values = pd.Series(np.concatenate([comp1, comp2]))   # Creamos distribución bimodal
sns.distplot(values, bins=100, color='k')
```

### Diagramas de dispersión

```python
macro = pd.read_csv('macrodata.csv')    # Cragamos el archivo cvs 'macrodata'
macro.head()
```

```python
data = macro[['cpi', 'm1', 'tbilrate', 'unemp']]     # elegimos cuatro variables
data.head()
```

```python
trans_data = np.log(data).diff().dropna() # Aplicamos diferencias de logaritmos, quitamos renglones NA
trans_data[-5:]  # Observamos los últimos cinco datos
```

```python
sns.regplot(x = 'm1', y = 'unemp', data=trans_data)    # Ajustamos una regresión y hacemos diagrama
plt.title('Changes in log %s versus log %s' % ('m1', 'unemp'))
```

```python
# También podemo hacer todas las combinaciones de diagramas con las distintas variables de 2 en 2
```

```python
sns.pairplot(trans_data, diag_kind='kde', plot_kws={'alpha': 0.2})       # Le agregamos KDE (kernel)
```

### Graficación de datos categóricos

```python
# Supenemos que nuestros datos se presentan con varias categorías
```

```python
sns.catplot(x='day', y='tip_pct', hue='time', col='smoker',
.....: kind='bar', data=tips[tips.tip_pct < 1])
```

```python
sns.catplot(x='day', y='tip_pct', row='time',    # Si queremos obtener frecuencias para todas
.....: col='smoker',                             # las combinaciones de categorías
.....: kind='bar', data=tips[tips.tip_pct < 1])
```

```python
# Otra alternativa:
```

```python
sns.catplot(x='tip_pct', y='day', kind='box',
.....: data=tips[tips.tip_pct < 0.5])             # Box muestra mediana quartile y outliers
```
